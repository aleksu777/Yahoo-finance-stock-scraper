```python

print(company_df.Symbol.nunique())
```

```python

## extracting only the energy sector firms!

comp_energy = company_df[(company_df['GICS Sector'] == 'Energy')]
comp_energy.head()

```

```python
comp_energy.Symbol.nunique()
```

```python
start_date = datetime.datetime(2020,7,4)
print(start_date)

end_date = datetime.datetime.today()

print(end_date)
```

```python
# Create new dataframe

stock_price = pd.DataFrame()

```

```python
# dummy table just to test for one company 

df = yf.download('APA', start = start_date, end = end_date )
```

```python
df
```

```python
for i in comp_energy.Symbol.unique():
  df = yf.download(i, start = start_date, end = end_date )

  df['Symbol'] = i

  stock_price = stock_price.append(df)

```

```python
stock_price
```

```python
# Illustrating it with a graph

import plotly.graph_objects as go


df = stock_price[stock_price['Symbol'] == 'APA']
fig = go.Figure(data=[go.Candlestick(x=df.index,
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])])

fig.show()
```

```python
# converting it to a csv file


stock_price.to_csv('test_prices.csv', index=True)
```
